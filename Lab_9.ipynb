# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def get_user_graph():
    graph = {}
    print("Enter graph details. Type 'done' when finished.")
    print("Format: Node1 Node2 Weight (e.g., A B 10)")

    while True:
        user_input = input("Enter edge (or 'done'): ")
        if user_input.lower() == 'done':
            break
        try:
            node1, node2, weight_str = user_input.split()
            weight = int(weight_str)

            if node1 not in graph:
                graph[node1] = {}
            graph[node1][node2] = weight
            if node2 not in graph:
                graph[node2] = {}
            graph[node2][node1] = weight

        except ValueError:
            print("Invalid input. Please use the format: Node1 Node2 Weight")

    return graph
user_graph = get_user_graph()

if user_graph:
    start_node = input("Enter the starting node for Dijkstra's algorithm: ")
    if start_node in user_graph:
        result = dijkstra(user_graph, start_node)
        print(f"\nShortest distances from node {start_node}:")
        for node in result:
            print(f"Distance from {start_node} to {node} = {result[node]}")
    else:
        print("Starting node not found in the graph.")
else:
    print("No graph data entered.")
```

```python
import heapq

def dijkstra(graph, start):
    distances = {node: float('infinity') for node in graph}
    distances[start] = 0
    priority_queue = [(0, start)]  # (distance, node)

    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)

        # If we've found a shorter path to current_node already, skip
        if current_distance > distances[current_node]:
            continue

        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight

            # If new path is shorter
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(priority_queue, (distance, neighbor))

    return distances
```
